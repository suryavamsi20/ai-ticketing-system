```python
import pandas as pd
import re
```

```python
df = pd.read_csv("../dataset/raw/aa_dataset-tickets-multi-lang-5-2-50-version.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.rename(columns={'body': 'text'}, inplace=True)
```

```python
df['text'].sample(5)
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text
```

```python
df['clean_text'] = df['text'].apply(clean_text)
```

```python
df[['text', 'clean_text']].sample(5)
```

```python
def remove_extra_spaces(text):
    return re.sub(r'\s+', ' ', text).strip()

df['clean_text'] = df['clean_text'].apply(remove_extra_spaces)
```

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

#Stopword Removal

def remove_stopwords(text):
    return " ".join(
        word for word in text.split() if word not in stop_words
    )

df['clean_text'] = df['clean_text'].apply(remove_stopwords)
```

```python
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    return " ".join(
        lemmatizer.lemmatize(word) for word in text.split()
    )

df['clean_text'] = df['clean_text'].apply(lemmatize_text)
```

```python
df[['text', 'clean_text']].sample(10)
```

```python
df.to_csv("../dataset/processed/tickets_cleaned.csv", index=False)

```
